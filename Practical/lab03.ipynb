#### Пункт 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
set_size = 1000
radius = 10
```

```python
coord_x = []
coord_y = []
while len(coord_x) != set_size:
    x = np.random.uniform(-radius, radius)
    y = np.random.uniform(-radius, radius)
    if x*x + y*y <= radius**2:
        coord_x.append(x)
        coord_y.append(y)
```

```python
circle_angle = np.linspace(0, 2 * np.pi, 150) 
circle_x = radius * np.cos(circle_angle)
circle_y = radius * np.sin(circle_angle)

plt.figure(figsize=(10, 10))
plt.plot(circle_x, circle_y, 'r')
plt.plot(coord_x, coord_y, "ko")
plt.title("При задании x и y")
plt.show()
```

```python
an_coord_x = []
an_coord_y = []
while len(an_coord_x) != set_size:
    angle = np.random.uniform(0, 2*np.pi)
    length = np.random.uniform(0, radius)
    an_coord_x.append(length * np.cos(angle))
    an_coord_y.append(length * np.sin(angle))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(an_coord_x, an_coord_y, 'ko')
plt.plot(circle_x, circle_y, 'r')
plt.title("При задании угла и расстояния")
plt.show()
```

#### Пункт 2

```python
print("Mean for x: ", np.mean(coord_x))
print("Mean for y: ", np.mean(coord_y))
```

```python
print("Var for x: ", np.var(coord_x))
print("Var for y: ", np.var(coord_y))
```

```python
dot_x = 20
dot_y = 0
```

```python
def dist(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
distances = []
for i in range(set_size):
    distances.append(dist(coord_x[i], coord_y[i], dot_x, dot_y))
distances.sort()
```

```python
kde_distances_y = gaussian_kde(distances).evaluate(distances)
```

```python
plt.figure(dpi=100)
plt.title("Плотность распределения между точками")
plt.plot(distances, kde_distances_y)
plt.hist(distances, bins=100, alpha=0.2, density=True)
plt.show()
```

```python
dist_1 = []
for c in range(set_size):
    i, j = np.random.randint(0, set_size - 1), np.random.randint(0, set_size - 1)
    dist_1.append(dist(coord_x[i], coord_y[i], coord_x[j], coord_y[j]))
dist_1.sort()

kde_dist_1_y = gaussian_kde(dist_1).evaluate(dist_1)
```

```python
plt.figure(dpi=100)
plt.title("Плотность распределения между точками")
plt.plot(dist_1, kde_dist_1_y)
plt.hist(dist_1, bins=100, alpha=0.2, density=True)
plt.show()
```
