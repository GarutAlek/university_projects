```python
from math import erf
def v_normal(x, u, o):
    return 0.5*(1 + erf((x - u)/np.sqrt(2 * o ** 2)))

normal_theory_x = np.linspace(-3, 3, 100)
normal_theory_y = []
for i in normal_theory_x:
    normal_theory_y.append(v_normal(i, 0, 1))
    
    
normal100_s = sorted(np.copy(normal100))
normal1000_s = sorted(np.copy(normal1000))
normal100_y = np.arange(0., 1., 0.01)
normal1000_y = np.arange(0., 1., 0.001)

plt.figure(figsize=(10, 10))
plt.plot(normal_theory_x, normal_theory_y, label="theory")
plt.plot(normal100_s, normal100_y, label="normal100")
plt.plot(normal1000_s, normal1000_y, label="normal1000")
plt.legend()
plt.title("normal")
plt.show()
```

```python
def v_exp(l, x):
    return 1 - np.exp(-l * x)

exp_theory_x = np.linspace(0, 10, 100)
exp_theory_y = v_exp(1, exp_theory_x)
exp100_s = sorted(np.copy(exp100))
exp1000_s = sorted(np.copy(exp1000))

exp100_y = np.arange(0., 1., 0.01)
exp1000_y = np.arange(0., 1., 0.001)

plt.figure(figsize=(10, 10))
plt.plot(exp_theory_x, exp_theory_y, label="theory")
plt.plot(exp100_s, exp100_y, label="exp100")
plt.plot(exp1000_s, exp1000_y, label="exp1000")
plt.legend()
plt.title("exponent")
plt.show()
```

### Задание 2

```python
side1 = 10
side2 = 30
```

```python
def distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)

def create_dist(value):
    dist = []
    for i in range(value):
        x1, y1 = np.random.uniform(0, side1), np.random.uniform(0, side2)
        x2, y2 = np.random.uniform(0, side1), np.random.uniform(0, side2)
        dist.append(distance(x1, y1, x2, y2))
    dist.sort()
    return dist

list100_y = create_dist(100)
list1000_y = create_dist(1000)
list10000_y = create_dist(10000)

length = np.sqrt(side1**2 + side2**2)
list100_x = np.linspace(0, length, 100)
list1000_x = np.linspace(0, length, 1000)
list10000_x = np.linspace(0, length, 10000)

plt.figure(figsize=(10, 10))
plt.plot(list100_x, list100_y, label="100")
plt.plot(list1000_x, list1000_y, label="1000")
plt.plot(list10000_x, list10000_y, label="10000")
plt.title("Распределение вероятностей")
plt.legend()
plt.show()
```

```python
print("mean: ", np.mean(list10000_y))
```

```python
from scipy.stats import gaussian_kde
kde_list100_y = gaussian_kde(list100_y).evaluate(list100_y)
kde_list1000_y = gaussian_kde(list1000_y).evaluate(list1000_y)
kde_list10000_y = gaussian_kde(list10000_y).evaluate(list10000_y)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(list100_y, kde_list100_y, label="100")
plt.plot(list1000_y, kde_list1000_y, label="1000")
plt.plot(list10000_y, kde_list10000_y, label="10000")
plt.legend()
plt.show()
```
