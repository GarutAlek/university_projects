<h4>Задание 1</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = np.random.uniform(size=(10, 2))
```

```python
plt.figure(dpi=150, figsize=(5, 5))
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.plot(matrix[:,0], matrix[:,1], 'ro')
```

<h4>Задание 2</h4>

```python
numbers = np.random.uniform(0, 10, 1000)
print('mean: ', np.mean(numbers))
print('var: ', np.var(numbers))
plt.hist(numbers, density=True, bins=100, histtype='stepfilled')
```
